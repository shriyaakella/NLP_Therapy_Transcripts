# Pre Processing

removing punctuation

```python
import string
def punctuation_count_before(txt):
    count= sum([1 for c in txt if c in string.punctuation])
    return 100*count/len(txt)
df3['punctaution_%']=df3['questionText'].apply(lambda x: punctuation_count_before(x))
df3.head()
```

```python
string.punctuation
def remove_punctuation(txt):
    txt_nopunct= "".join([c for c in txt if c not in string.punctuation])
    return txt_nopunct

df3['questionText']= df3['questionText'].apply(lambda x: remove_punctuation(x))
df3.head()
```

```python
def punctuation_count_after(txt):
    count= sum([1 for c in txt if c in string.punctuation])
    return 100*count/len(txt)
df3['punctaution_after%']=df3['questionText'].apply(lambda x: punctuation_count_after(x))
df3.head()
```

removing stopwords

```python
import re
def tokenize(txt):
    tokens= re.split('\W+',txt)
    return tokens
df3['questionText']=df3['questionText'].apply(lambda x: tokenize(x.lower())) #using lower to not differntiate b/w uppercase and lower case
```

```python
import nltk
#nltk.download('stopwords')
stopwords= nltk.corpus.stopwords.words('english')
```

```python
def remove_stopwords(txt_tokenized):
    tokens= [word for word in txt_tokenized if word not in stopwords]
    return tokens
df3['tokens']=df3['questionText'].apply(lambda x:remove_stopwords(x))
df3['questionText']=df3['tokens'].apply(lambda x:" ".join(x))
df3.head()
```

Lemmatizing words

```python
import nltk
#nltk.download('wordnet')
wn= nltk.WordNetLemmatizer()
ps= nltk
```

```python
def lemmatization(token_txt):
    text= [wn.lemmatize(word) for word in token_txt]
    return text
df3['lemmatized']=df3['tokens'].apply(lambda x:lemmatization(x))

df3['questionText']=df3['lemmatized'].apply(lambda x:" ".join(x))
df3.head()
```

# Model

```python
train = df3[df3["split"] == "train"]
test = df3[df3["split"] == "test"]
val = df3[df3["split"] == "val"]
val.shape
```

```python
df3['target'].value_counts()
```

```python
X_train = train['questionText']
y_train = train['target']
X_test = test['questionText']
y_test = test['target']
X_val = val['questionText']
y_val = val['target'] 
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer

C_vect = CountVectorizer(min_df=7, ngram_range=(1,3)).fit(X_train)
TF_vect = TfidfVectorizer(min_df=7).fit(X_train) 

(len(C_vect.get_feature_names()),len(TF_vect.get_feature_names()))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import SVC
from sklearn.metrics import roc_auc_score
from sklearn.metrics import classification_report, confusion_matrix

X_train_vectorized_TF = TF_vect.transform(X_train)
X_train_vectorized_C = C_vect.transform(X_train)
```

```python
clf = LogisticRegression(C=1)
clf.fit(X_train_vectorized_TF, y_train)

predictions = clf.predict(TF_vect.transform(X_test))

print('AUC: ', roc_auc_score(y_test, predictions))
print(classification_report(y_test, predictions))

cm = confusion_matrix(y_test, predictions)


ax= plt.subplot()
sns.heatmap(cm, annot=True, ax = ax); 

# labels, title and ticks
ax.set_xlabel('Predicted labels');
ax.set_ylabel('True labels'); 
ax.set_title('Confusion Matrix');
```
